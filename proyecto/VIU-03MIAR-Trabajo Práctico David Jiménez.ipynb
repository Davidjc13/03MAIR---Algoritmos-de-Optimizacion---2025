Respuesta:

Estamos en un problema de minimización, ya que la función objetivo trata de minimizar la diferencia entre el resultado de la expresión y el valor objetivo.

$$f_{objetivo} = eval(expresion) - objetivo$$


Tratamos de obtener que la diferencia sea 0, aunque como estamos trabajando con python se establecerá una tolerancia.

Diseña un algoritmo para resolver el problema por fuerza bruta

````python
import itertools

def fuerzabruta(objetivo):
    """
    Resuelve el problema propuesto arriba por fuerza bruta
    """

    #Nuestros dígitos serán del 1 al 9
    digitos = '123456789'
    #Definimos los 4 operadores
    operadores = ['+', '-', '*', '/']
    #Creamos una lista donde almacenaremos las soluciones
    soluciones = []
    
    #Seleccionamos 5 dígitos sin repetición
    for digitos_permutados in itertools.permutations(digitos, 5):
        #Seleccionamos 4 operadores sin repetición
        for operadores_permutados in itertools.permutations(operadores, 4):

            expresion = digitos_permutados[0]

            #Construimos la expresión algebraica (dígito operador dígito operador ...)
            for i in range(4):
                expresion += operadores_permutados[i] + digitos_permutados[i+1]

            #Manejamos la excepción al dividir por cero
            try:
                #Se usa una tolerancia y no una igualdad para evitar problemas,
                #si la expresión es igual al objetivo añadimos la solución 
                if abs(eval(expresion) - objetivo) < 1e-6:
                    soluciones.append(expresion)
            except ZeroDivisionError:
                continue


    return soluciones




#PROGRAMA PRINCIPAL

numero_objetivo = 22
print("Soluciones para el objetivo", numero_objetivo, ":", fuerzabruta(numero_objetivo))
````

Calcula la complejidad del algoritmo por fuerza bruta

Respuesta: 

La complejidad del algoritmo se determina multiplicando el número de permutaciones de 5 dígitos de un conjunto de 9, por el número de permutaciones de los 4 operadores, resultando en un total de 362.880 expresiones calculado arriba. Por lo tanto en términos de complejidad el algoritmo es factorial $O(n!)$ respecto a los parámetros número de dígitos y número de operadores.

(*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

Respuesta: El siguiente algoritmo mejora al de fuerza bruta porque agrega condiciones de poda sin explorar todas las combinaciones completas de forma exhaustiva. Aunque en el peor de los casos la complejidad sigue siendo factorial, permite reducir el número de evaluaciones mediante la poda lo que resulta en una búsqueda más eficiente.

````python
def resuelve_backtracking(objetivo):
    """
    Resuelve el problema descrito en el enunciado mediante backtracking (vuelta atrás)
    """
    soluciones = []
    

    def backtrack(expresion, digitos_dispo, operadores_dispo):

        #Si la expresión alcanza una longitud de 9 ya hemos terminado
        if len(expresion) == 9:

            #Manejamos la excepción al  dividir por cero
            try:
                #Agregamos solucion si es igual al número objetivo
                if abs(eval(expresion) - objetivo) < 1e-6:
                    soluciones.append(expresion)
            except ZeroDivisionError:
                pass
            return
        

        #                       TÉCNICA DE PODA
        #----------------------------------------------------------------
        if len(expresion) > 0 and len(expresion) % 2 == 1:
            try:
                valor_actual = eval(expresion)
            except ZeroDivisionError:
                #Si hay división por cero, podamos
                return
            
            suma_restante = sum(int(d) for d in digitos_dispo)

            #Aplicamos poda si con los dígitos restantes no se puede alcanzar el valor objetivo (sumando)
            if valor_actual < objetivo and valor_actual + suma_restante < objetivo:
                return
            ##Aplicamos poda si con los dígitos restantes no se puede alcanzar el valor objetivo (restando)
            if valor_actual > objetivo and valor_actual - suma_restante > objetivo:
                return
        #----------------------------------------------------------------


        #Si la longitud de la expresión es par esperamos un dígito
        if len(expresion) % 2 == 0:

            for i in range(len(digitos_dispo)):
                digito = digitos_dispo[i]
                #Aplicamos la técnica vuelta atrás
                backtrack(expresion + digito, digitos_dispo[:i] + digitos_dispo[i+1:], operadores_dispo)
        else:
            #Si la longitud de la expresión es impar esperamos un operador
            for i in range(len(operadores_dispo)):
                operador = operadores_dispo[i]
                #Aplicamos la técnica vuelta atrás
                backtrack(expresion + operador, digitos_dispo, operadores_dispo[:i] + operadores_dispo[i+1:])

    backtrack("", list("123456789"), ['+', '-', '*', '/'])
    return soluciones




#PROGRAMA PRINCIPAL

valor_objetivo = 4
soluciones = resuelve_backtracking(valor_objetivo)
print("Soluciones para el objetivo", valor_objetivo, ":", soluciones)
````

(*)Calcula la complejidad del algoritmo

Respuesta: 
La complejidad del algoritmo de backtracking o técnica de vuelta atrás depende en gran medida de la implementación de las posibles podas, para el caso planteado arriba, la complejidad se podría reducir incorporando alguna heurística o técnida de poda adicional. En el peor de los casos este algoritmo iguala en complejidad al algoritmo de fuerza bruta, teniendo así una complejidad factorial $O(n!)$ aunque suele ser más eficiente a la hora de ejecutarse, mejorando de forma sustancial al algoritmo por fuerza bruta.
